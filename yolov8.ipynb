```python
import os
import shutil
import re

# Path to your images folder
source_folder = "/content/drive/My Drive/mam/"
benign_folder = os.path.join(source_folder, "benign")
malignant_folder = os.path.join(source_folder, "malignant")
no_detection_folder = os.path.join(source_folder, "no_detection")

os.makedirs(benign_folder, exist_ok=True)
os.makedirs(malignant_folder, exist_ok=True)
os.makedirs(no_detection_folder, exist_ok=True)

# Your log lines — replace this list with reading from your actual log file if needed
log_lines = [
    "image 315/322 /content/drive/My Drive/mam/mdb315.png: 640x640 (no detections), 7.4ms",
    "image 316/322 /content/drive/My Drive/mam/mdb316.png: 640x640 1 benign, 7.4ms",
    "image 317/322 /content/drive/My Drive/mam/mdb317.png: 640x640 1 benign, 7.4ms",
    "image 322/322 /content/drive/My Drive/mam/mdb322.png: 640x640 1 malignant, 7.4ms",
    # Add all your actual log lines here
]

results = {}

# Regex patterns to find file name and detection type
pattern = re.compile(r"/mam/(mdb\d+\.png):.*?(no detections|benign|malignant)")

for line in log_lines:
    match = pattern.search(line)
    if match:
        img_name = match.group(1)
        detection = match.group(2)
        # Normalize detection labels
        if detection == "no detections":
            results[img_name] = "no_detection"
        else:
            results[img_name] = detection

# Now move files according to parsed results
for img_name, label in results.items():
    src_path = os.path.join(source_folder, img_name)
    if not os.path.exists(src_path):
        print(f"Image {img_name} not found!")
        continue

    if label == "benign":
        dst_path = os.path.join(benign_folder, img_name)
    elif label == "malignant":
        dst_path = os.path.join(malignant_folder, img_name)
    else:  # no_detection
        dst_path = os.path.join(no_detection_folder, img_name)

    shutil.move(src_path, dst_path)
    print(f"Moved {img_name} to {label} folder.")
```

```python
import os
import shutil
from ultralytics import YOLO

# Load your trained YOLO model
model = YOLO("/content/drive/My Drive/trained/cancer_cell2/weights/best.pt")

# Folders setup
source_folder = "/content/drive/My Drive/mam/"
benign_folder = os.path.join(source_folder, "benign")
malignant_folder = os.path.join(source_folder, "malignant")
no_detection_folder = os.path.join(source_folder, "no_detection")

os.makedirs(benign_folder, exist_ok=True)
os.makedirs(malignant_folder, exist_ok=True)
os.makedirs(no_detection_folder, exist_ok=True)

# Map class IDs to labels (adjust these based on your model's classes)
class_map = {0: "benign", 1: "malignant"}

# Get all images in the source folder
image_files = [f for f in os.listdir(source_folder) if f.endswith('.png')]

for img_file in image_files:
    img_path = os.path.join(source_folder, img_file)
    results = model(img_path)

    detected_classes = []
    for r in results:
        if r.boxes is not None and len(r.boxes) > 0:
            for box in r.boxes:
                cls_id = int(box.cls[0])
                detected_classes.append(class_map.get(cls_id, "unknown"))

    if not detected_classes:
        dst_folder = no_detection_folder
    elif "malignant" in detected_classes:
        dst_folder = malignant_folder
    else:
        dst_folder = benign_folder

    dst_path = os.path.join(dst_folder, img_file)
    shutil.copy(img_path, dst_path)
    print(f"Copied {img_file} to {dst_folder}")

```

```python
import os
import shutil
import time
from ultralytics import YOLO

model = YOLO("/content/drive/My Drive/trained/cancer_cell2/weights/best.pt")

source_folder = "/content/drive/My Drive/mam/"
benign_folder = os.path.join(source_folder, "benign")
malignant_folder = os.path.join(source_folder, "malignant")
no_detection_folder = os.path.join(source_folder, "no_detection")

os.makedirs(benign_folder, exist_ok=True)
os.makedirs(malignant_folder, exist_ok=True)
os.makedirs(no_detection_folder, exist_ok=True)

class_map = {0: "benign", 1: "malignant"}

image_files = [f for f in os.listdir(source_folder) if f.endswith('.png')]

start_time = time.time()  # Start timing

for img_file in image_files:
    img_path = os.path.join(source_folder, img_file)
    results = model(img_path)

    detected_classes = []
    for r in results:
        if r.boxes is not None and len(r.boxes) > 0:
            for box in r.boxes:
                cls_id = int(box.cls[0])
                detected_classes.append(class_map.get(cls_id, "unknown"))

    if not detected_classes:
        dst_folder = no_detection_folder
    elif "malignant" in detected_classes:
        dst_folder = malignant_folder
    else:
        dst_folder = benign_folder

    dst_path = os.path.join(dst_folder, img_file)
    shutil.copy(img_path, dst_path)

end_time = time.time()  # End timing

total_time = end_time - start_time
avg_time_per_image = total_time / len(image_files)

print(f"Total inference time for {len(image_files)} images: {total_time:.2f} seconds")
print(f"Average time per image: {avg_time_per_image:.2f} seconds")
```

```python
import time
from ultralytics import YOLO
from pathlib import Path
import os

# ✅ Load model
model = YOLO("/content/drive/MyDrive/yolo_test_result/trained/cancer_cell2/weights/best.pt")

# ✅ Define input and output paths
input_dir = "/content/drive/My Drive/new"
save_path = Path("/content/drive/My Drive/new_result")
save_path.mkdir(parents=True, exist_ok=True)

# ✅ Start time tracking
start_time = time.time()

# ✅ Run predictions on folder
results = model.predict(
    source=input_dir,
    conf=0.25,
    save=True,
    save_dir=save_path
)

# ✅ Initialize counters
malignant_count = 0
benign_count = 0
no_detection_count = 0

# ✅ Process each result
for r in results:
    labels = r.boxes.cls.tolist() if r.boxes else []

    has_malignant = any(label == 1 for label in labels)  # YOLO class 1 = malignant
    has_benign = any(label == 0 for label in labels)     # YOLO class 0 = benign

    if has_malignant:
        malignant_count += 1
    elif has_benign:
        benign_count += 1
    else:
        no_detection_count += 1

# ✅ Time tracking
end_time = time.time()
total_time = end_time - start_time
total_images = len(results)
avg_time = total_time / total_images if total_images > 0 else 0

# ✅ Detection Summary
print("\n📊 Detection Summary:")
print(f"🟥 Malignant Images: {malignant_count}")
print(f"🟩 Benign Images: {benign_count}")
print(f"⚪ No Detections: {no_detection_count}")
print(f"📸 Total Images: {total_images}")

# ✅ Time Summary
print("\n⏱️ Inference Time:")
print(f"🕒 Total Time Taken: {total_time:.2f} seconds")
print(f"🕐 Average Time per Image: {avg_time:.4f} seconds")

# ✅ Evaluation Summary
print("\n📌 Evaluation Summary:")
print(f"✅ Model Used: {model.model.name}")
print(f"🎯 Confidence Threshold: 0.25")
print(f"📁 Predictions Saved To: {save_path}")
```
